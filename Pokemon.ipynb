```python
%matplotlib inline
```

```python
import pandas as pd
pokemon=pd.read_csv("E:\\Data Science Research\\Pokemon.csv",encoding='ISO-8859-1')
```

```python
pokemon.head(1)
```

```python
# 10.All top 10   water type pokemon that  rank are higher on HP
pokemon[pokemon.Type=='WATER'].head(10).groupby(pokemon['Name']).agg({'HP':'sum'}).sort_values(['HP'],ascending=False).plot(kind='bar',color='y')#.head(10).groupby([pokemon['Name'],pokemon['Type']]).agg({'HP':'sum'}).sort_values(['Name'],ascending=False).plot(kind='bar',label='Special Defense',color='r').legend()
```

```python
# 9.All type of pokemon that higher special defense
pokemon['Special Defense'].groupby([pokemon['Type']]).sum().sort_values().plot(kind='bar',label='Special Defense',color='r').legend()
```

```python
# 8.All type of pokemon that higher defense
pokemon.Defense.groupby([pokemon['Type']]).sum().sort_values().plot(kind='bar',label='Defense').legend()
```

```python
# 7.List top 10   pokemon with type that rank of higher HP.
pokemon.head(10).groupby([pokemon['Name'],pokemon['Type']]).agg({'HP':'sum'}).sort_values(['HP'],ascending=False).plot(kind='bar',color='c')
```

```python
# 6.List top 10  pokemon that rank of higher HP.
pokemon.head(10).groupby([pokemon['Name']]).agg({'HP':'sum'}).sort_values(['HP'],ascending=False).plot(kind='bar',color='g')
```

```python
# 5.List of pokemon that rank of higher HP.
pokemon.groupby([pokemon['Type']]).agg({'HP':'sum'}).sort_values(['HP']).plot(kind='bar',color='g')
```

```python
# 4.List of pokemon that rank of attack.
pokemon.groupby([pokemon['Type']]).agg({'Special Attack':'sum'}).sort_values(['Special Attack']).plot(kind='bar',color='c')
```

```python
# 3.Who's pokemon that attack high
pokemon.groupby([pokemon['Type']]).agg({'Attack':'sum'}).sort_values(['Attack']).plot(kind='bar',color='r')
```

```python
# 2.All type of pokemon speed
pokemon.groupby([pokemon['Type']]).agg({'Speed':'sum'}).sort_values(['Speed']).plot(kind='bar')
```

```python
# 1.All type of pokemon count
pokemon.groupby([pokemon['Type']])['Type'].count().sort_values().plot(kind='bar',color='c')
```
